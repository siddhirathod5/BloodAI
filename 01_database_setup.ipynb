Search Function

```python
def find_donor(blood_group, city):

    query = f"""
    SELECT *
    FROM donors
    WHERE blood_group='{blood_group}'
    AND city='{city}'
    """

    return pd.read_sql(query, engine)
```

```python
find_donor("O+", "Ahmedabad")
```

```python
query = """
SELECT
blood_group,
COUNT(*) AS total
FROM donors
GROUP BY blood_group
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
city,
COUNT(*) AS total
FROM donors
GROUP BY city
"""

pd.read_sql(query, engine)
```
